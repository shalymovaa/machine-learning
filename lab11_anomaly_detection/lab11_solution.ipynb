```python
from mat4py import loadmat
import matplotlib.pyplot as plt
import numpy as np
import math
import random
data = loadmat('ex8data1.mat')
X=np.array(data['X'])
for i in range(len(X)):
    plt.plot(X[i][0],X[i][1],'rx')
```

```python
mu=[]
for i in range(len(X[0])):
    m=np.mean(X[:,i])
    mu.append(m)
print(mu)
```

```python
sigma=[]
for i in range(len(X[0])):
    s=np.std(X[:,i])
    sigma.append(s)
print(sigma)
```

```python
s=(1/len(X))*np.sum((X[:,0]-mu[0])**2)
print(s)
```

```python
from numpy import random
Xval=np.array(data['Xval'])
p=np.zeros((len(Xval),2))
for i in range(len(Xval)):
    for j in range(len(Xval[0])):
        p[i][j]=(1/math.sqrt(2*math.pi*(sigma[j]**2)))*math.exp(-0.5*((Xval[i][j]-mu[j])**2)/(sigma[j]**2))
```

```python
y=[]
for i in range(len(p)):
    p1=np.prod(p[i])
    y.append(p1)
```

```python
e=8.99e-05
cnt=0
ind=[]
for i in range(len(y)):
    if y[i]<e:
        cnt=cnt+1
        ind.append(i)
print(cnt)
print(ind)
```

```python
y1=np.array(data['yval'])
cnt1=0
ind1=[]
for i in range (len(y1)):
    if y1[i]==1:
        cnt1=cnt1+1
        ind1.append(i)
print(cnt1)
print(ind1)
        
```

```python
for i in range(len(Xval)):
    if i in ind:
        plt.plot(Xval[i][0],Xval[i][1],'ro')
    else:    
        plt.plot(Xval[i][0],Xval[i][1],'bo')
```
