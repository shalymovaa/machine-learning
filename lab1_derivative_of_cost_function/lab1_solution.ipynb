```python

import matplotlib.pyplot as plt
import csv
import numpy as np
path='C:\\Users\shaly\Desktop\dataset.txt'

X = []
Y = []
H=[]

with open(path, 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=',')
    for ROWS in plotting:
        X.append(float(ROWS[0]))
        Y.append(float(ROWS[1]))
        
for i in (X):
    h=0.01+0.8*i
    H.append(h)
        
plt.plot(X, Y, 'x',color='black')
plt.plot(X, H)
```

```python
tet0=np.linspace(-50,50,100)
tet1=np.linspace(-50,50,100)
m=97
Js=np.zeros((100,100))
for j in range(len(tet0)):
    for k in range (len(tet1)):
        J=[]
        for i in range (len(X)):
            ms=(tet0[j]+tet1[k]*X[i]-Y[i])**2
            J.append(ms)
        s=1/(2*m)*sum(J)
        Js[j][k]=s
print(Js)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(tet0,tet1,Js, 200, cmap='binary')
ax.set_xlabel('tetta0')
ax.set_ylabel('tetta1')
ax.set_zlabel('J');
ax.scatter3D(0.01, 0.8, 26.593283409687096, cmap='Greens')
```
