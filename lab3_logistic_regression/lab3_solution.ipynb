```python
#lambda=0
import matplotlib.pyplot as plt
import csv
import numpy as np
import math
path='C:\\Users\shaly\Desktop\ex2data2.txt'
tet0=4
tet1=1
tet2=2
tet3=0
tet4=0
a=0.0001
n=200000
m=118
lam=0
X = []
Y = []
Z = []
sj=[]
cnt=0

with open(path, 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=',')
    for ROWS in plotting:
        X.append(float(ROWS[0]))
        Y.append(float(ROWS[1]))
        Z.append(float(ROWS[2]))
s1=max(X)
s2=max(Y)
average_x=sum(X)/len(X)
average_y=sum(Y)/len(Y)

for i in range(len(X)):
    X[i]=(X[i]-average_x)/s1
    Y[i]=(Y[i]-average_y)/s2
    
while cnt!=n:
    cnt=cnt+1
    dtet0=[]
    dtet1=[]
    dtet2=[]
    dtet3=[]
    dtet4=[]
    for j in range (len(X)):
        ms=(1/(1+math.exp(-tet0-tet1*X[j]-tet2*Y[j]-tet3*X[j]**2-tet4*Y[j]**2))-Z[j])*1
        ms1=(1/(1+math.exp(-tet0-tet1*X[j]-tet2*Y[j]-tet3*X[j]**2-tet4*Y[j]**2))-Z[j])*X[j]
        ms2=(1/(1+math.exp(-tet0-tet1*X[j]-tet2*Y[j]-tet3*X[j]**2-tet4*Y[j]**2))-Z[j])*Y[j]
        ms3=(1/(1+math.exp(-tet0-tet1*X[j]-tet2*Y[j]-tet3*X[j]**2-tet4*Y[j]**2))-Z[j])*X[j]**2
        ms4=(1/(1+math.exp(-tet0-tet1*X[j]-tet2*Y[j]-tet3*X[j]**2-tet4*Y[j]**2))-Z[j])*Y[j]**2
        dtet0.append(ms)
        dtet1.append(ms1)
        dtet2.append(ms2)
        dtet3.append(ms3)
        dtet4.append(ms4)
        
    temp0=tet0-a/m*sum(dtet0)
    temp1=tet1-a/m*sum(dtet1)+(lam/m)*tet1
    temp2=tet2-a/m*sum(dtet2)+(lam/m)*tet2
    temp3=tet3-a/m*sum(dtet3)+(lam/m)*tet3
    temp4=tet4-a/m*sum(dtet4)+(lam/m)*tet4
    tet0=temp0
    tet1=temp1
    tet2=temp2
    tet3=temp3
    tet4=temp4
    
    J=[]
    for k in range (len(X)):
        mj=Z[k]*math.log(1/(1+math.exp(-tet0-tet1*X[k]-tet2*Y[k]-tet3*X[k]**2-tet4*Y[k]**2)))+(1-Z[k])*(math.log(1-(1/(1+math.exp(-tet0-tet1*X[k]-tet2*Y[k]-tet3*X[k]**2-tet4*Y[k]**2)))))
        J.append(mj)
    s=-1/(m)*sum(J)+lam/(2*m)*(tet0**2+tet1**2+tet2**2+tet3**2+tet4**2)
    sj.append(s) 
print(tet0,tet1,tet2,tet3,tet4) 
plt.plot(np.linspace(0,n-1,n),sj)
```

```python
import numpy as np

def quadratic_eq(b: float, c: float):
    if b**2 - 4 * c < 0:
        return None, None
    x1 = (-b + np.sqrt(b**2 - 4 * c)) / 2
    x2 = (-b - np.sqrt(b**2 - 4 * c)) / 2
    ans = [x1, x2]
    ans.sort()
    return ans

def get_decision_boundary(N: int, X_min: float, X_max: float, ths):
    ys_lower = []
    ys_upper = []
    b = ths[2] / ths[4]
    xs = np.linspace(X_min, X_max, N)
    for x1 in xs:
        c = (ths[0] + x1 * ths[1] + x1**2 * ths[3]) / ths[4]
        y21, y22 = quadratic_eq(b, c)
        # if y21 is not None:
        ys_lower.append(y21)
        ys_upper.append(y22)

    return xs, ys_lower, ys_upper
    
```
